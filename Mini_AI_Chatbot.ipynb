## While looop, and if-elif

```python
while True:
    print("\n--------------------------")
    print("--- Questions Menu ---")
    print("1. Resume language?")
    print("2. How to apply?")
    print("3. Location?")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("\nANSWER:")
        print("English resumes are preferred.")
    elif choice == "2":
        print("\nANSWER:")
        print("Apply through our website.")
    elif choice == "3":
        print("\nANSWER:")
        print("Saudi Arabia, Makkah.")
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice.")
```

# Using a Dictionary

```python
responses = {
    "1": "English resumes are preferred.",
    "2": "Apply through our website.",
    "3": "Saudi Arabia, Makkah.",
}
```

```python
while True:
    print("\n--- Menu ---")
    print("1. Resume language?")
    print("2. How to apply?")
    print("3. Location?")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "4":
        print("Goodbye!")
        break
    elif choice in responses:
        print(responses[choice])
    else:
        print("Invalid choice.")
```

# Text-Based Chatbot (User types questions)

```python
while True:
    user_input = input("Ask me something (or type 'exit'): ").lower()

    if user_input == "exit":
        print("Goodbye!")
        break
    elif "resume" in user_input:
        print("English resumes are preferred.")
    elif "apply" in user_input:
        print("Apply through our website.")
    elif "location" in user_input:
        print("Saudi Arabia, Makkah.")
    else:
        print("Sorry, I don't understand.")
```

# Function-Based Version

```python
def show_menu():
    print("\n1. Resume language")
    print("2. How to apply")
    print("3. Location")
    print("4. Exit")
```

```python
def get_answer(choice):
    answers = {
        "1": "English resumes are preferred.",
        "2": "Apply through our website.",
        "3": "Saudi Arabia, Makkah."
    }
    return answers.get(choice, "Invalid choice.")
```

```python
while True:
    show_menu()
    choice = input("Enter your choice: ")

    if choice == "4":
        print("Goodbye!")
        break

    print(get_answer(choice))
```
